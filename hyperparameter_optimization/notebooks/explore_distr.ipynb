```python
import math
```

```python
math.exp(10)
```

```python
def qlognormal(mu, sigma, q, size):
    sampling = [int(math.exp(np.random.normal(mu, sigma))/q)*q for i in range(size)]
    return sampling
```

```python
sampling = qlognormal(math.log(10), math.log(1000), 1, 10000)
```

```python
from hyperopt import hp
from hyperopt.pyll import scope
from hyperopt.pyll.stochastic import sample
```

```python
space = scope.int(hp.qlognormal('loguniform', 1, 0.5, 1))
```

```python
sampling = [sample(space) for i in range(10000)]
```

```python
sns.displot(sampling)
```

```python
min(sampling)
```
